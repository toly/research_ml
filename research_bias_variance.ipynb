```python
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor

import matplotlib.pyplot as plt

%matplotlib inline
```

# test linear

```python
n = 1000
x = np.random.random(n) * 2 + 2
y = x * 5 + 2 + np.random.randn(n) * 3
```

```python
plt.scatter(x, y, alpha=0.1)
plt.show()
```

```python
x_train = x[:500]
y_train = y[:500]
x_test = x[500:]
y_test = y[500:]
```

```python
model = LinearRegression()
model.fit(np.expand_dims(x_train, 1), y_train)
```

```python
y_test_predict = model.predict(np.expand_dims(x_test, 1))
```

```python
plt.scatter(y_test, y_test_predict, alpha=0.5)
plt.show()
```

```python
errors = y_test - y_test_predict
```

```python
plt.hist(errors, bins=100)
plt.show()
```

# test polynomial

```python
n = 1000
x = np.random.random(n) * 10 - 5
y = 2*x**2 + x * 5 - 2 + np.random.randn(n) * 3
```

```python
plt.scatter(x, y, alpha=0.1)
plt.show()
```

```python
x_train = x[:500]
y_train = y[:500]
x_test = x[500:]
y_test = y[500:]
```

```python
model = LinearRegression()
model.fit(np.expand_dims(x_train, 1), y_train)
```

```python
y_test_predict = model.predict(np.expand_dims(x_test, 1))
```

```python
plt.scatter(y_test, y_test_predict, alpha=0.5)
plt.show()
```

```python
errors = y_test - y_test_predict
```

```python
plt.hist(errors, bins=100)
plt.show()
```

```python
errors.mean()
```

```python
plt.scatter(x_test, errors)
plt.show()
```

```python
y_train_predict = model.predict(np.expand_dims(x_train, 1))
```

```python
plt.scatter(y_train, y_train_predict, alpha=0.5)
plt.show()
```

```python
errors_train = y_train - y_train_predict
```

```python
plt.hist(errors_train, bins=100)
plt.show()
```

```python
errors_train.mean()
```

```python
plt.scatter(x_train, errors_train)
plt.show()
```
